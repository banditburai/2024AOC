ok our next task is to calculate how many times each number from our left list appears in the right list.
So let's brainstorm options on how to achieve that. We can use a set for the left list and then iterate over the right list and count how many times each number appears in the right list. We can use a dictionary to save the counts. And then we can iterate over list 1 and multiple to get the similarity score.

```python
left_set = set(left_sorted)
```

```python
matches = {num: 0 for num in left_set}  # initialize all counts to 0
for num in right_sorted:
    if num in left_set:
        matches[num] += 1
```

```python
# show non zero matches
non_zero_matches = {k: v for k, v in matches.items() if v > 0}
```

```python
similarity_score = sum(num * count for num, count in non_zero_matches.items())
similarity_score
```

This code is creating a dictionary using a dictionary comprehension. Let's break it down:
it's equivalent to:
matches = {}
for num in left_set:
    matches[num] = 0

```python
# Define code snippets
code_init_set = """left_set = set(left_sorted)"""

code_count_matches = """matches = {num: 0 for num in left_set}  # initialize all counts to 0
for num in right_sorted:
    if num in left_set:
        matches[num] += 1"""

code_filter_matches = """# show non zero matches
non_zero_matches = {k: v for k, v in matches.items() if v > 0}"""

code_similarity = """similarity_score = sum(num * count for num, count in non_zero_matches.items())"""

# Add content to page
home.add(
    H2("Calculating Similarity Score"),
    P("""After finding the minimum total distance, we need to calculate a similarity score. 
        This score is based on how many times each number from the left list appears in the right list."""),
    
    Div(
        H3("Step 1: Create Set for Efficient Lookup"),
        PrismCode(code_init_set),
        Ul(
            Li("✓ Converts list to set for O(1) lookups"),
            Li("✓ Removes duplicates automatically"),
            Li("✓ Memory efficient for large datasets"),
            Li("⚠ Order is not preserved (but not needed here)"),
        ),
        _class="solution"
    ),
    
    Div(
        H3("Step 2: Count Matching Numbers"),
        PrismCode(code_count_matches),
        Ul(
            Li("✓ Initialize counts to zero for all left numbers"),
            Li("✓ Efficiently count occurrences in right list"),
            Li("✓ Only counts numbers that exist in left list"),
            Li("⚠ Requires additional memory for dictionary"),
        ),
        _class="solution"
    ),
    
    Div(
        H3("Step 3: Filter Non-Zero Matches"),
        PrismCode(code_filter_matches),
        Ul(
            Li("✓ Removes numbers with zero matches"),
            Li("✓ Creates cleaner dictionary for final calculation"),
            Li("✓ Makes debugging easier"),
            Li("⚠ Creates new dictionary in memory"),
        ),
        _class="solution"
    ),
    
    Div(
        H3("Step 4: Calculate Similarity Score"),
        PrismCode(code_similarity),
        Ul(
            Li("✓ Multiplies each number by its count"),
            Li("✓ Sums all products for final score"),
            Li("✓ Uses generator expression for memory efficiency"),
            Li("✓ Only processes numbers that had matches"),
        ),
        _class="solution"
    ),
    
    Div(
        H3("Why This Approach?"),
        P("""This multi-step approach provides an efficient way to calculate similarity:"""),
        Ul(
            Li("Using a set makes lookups very fast"),
            Li("Dictionary keeps track of counts without multiple passes"),
            Li("Filtering non-zero matches reduces final calculation work"),
            Li("Generator expression in final sum keeps memory usage low"),
        ),
        _class="recommendation"
    )
)
```
