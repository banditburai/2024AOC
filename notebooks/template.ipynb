```python
def PrismCode(code, language="python"):
    """Create a Prism-highlighted code block
    
    Args:
        code (str): The code to highlight
        language (str, optional): Programming language. Defaults to "python".
    """
    return Pre(
        Code(
            code,
            _class=f"language-{language}"
        ),
        _class=f"language-{language}"
    )
```
